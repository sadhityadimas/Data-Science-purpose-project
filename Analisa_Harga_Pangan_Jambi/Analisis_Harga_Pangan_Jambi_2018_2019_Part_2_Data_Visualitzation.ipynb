```python
#histogram
def histogram_all(data_frame, date, save=False):
    list_ax2 = [0, 'ax1', 'ax2', 'ax3', 'ax4', 'ax5', 'ax6', 'ax7', 'ax8', 'ax9', 'ax10']
    columns_df2 = [0, 'Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah', 'Cabai Rawit', 
                   'Daging Ayam', 'Daging Sapi', 'Gula Pasir', 'Minyak Goreng', 'Telur Ayam']
    if date == '2018':
        fig = plt.figure(figsize=(20, 10))
        for i in range(1, 11):
            list_ax2[i] = plt.subplot(2, 5, i)
            plt.hist(data_frame['2018'][columns_df2[i]])#, bins=range(len(df_pivot['Bawang Merah'])))
            #list_ax[i].set_xticks(range(len(df_pivot)))
            #list_ax[i].set_xticklabels(df_pivot['Hari'])
            plt.title(columns_df2[i])
            #plt.xticks(rotation=90)
        plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
        plt.tight_layout()
        fig.suptitle(date, fontsize=16, y=1.03)
        if save == True:
            plt.savefig('tes2.png')
        plt.show()
        return
    if date == '2019':
        fig = plt.figure(figsize=(20, 10))
        for i in range(1, 11):
            list_ax2[i] = plt.subplot(2, 5, i)
            plt.hist(data_frame['2019'][columns_df2[i]])#, bins=range(len(df_pivot['Bawang Merah'])))
            #list_ax[i].set_xticks(range(len(df_pivot)))
            #list_ax[i].set_xticklabels(df_pivot['Hari'])
            plt.title(columns_df2[i])
            #plt.xticks(rotation=90)
        plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
        plt.tight_layout()
        fig.suptitle(date, fontsize=16, y=1.03)
        if save == True:
            plt.savefig('tes2.png')
        plt.show()
        return
    year = ['2018', '2019']
    if date == 'all':
        for i in range(len(year)):
            figure = plt.figure(figsize=(20, 10))
            for x in range(1, 11):
                list_ax2[x] = plt.subplot(2, 5,x)
                plt.hist(data_frame[year[i]][columns_df2[x]])#, bins=range(len(df_pivot['Bawang Merah'])))
                #list_ax[i].set_xticks(range(len(df_pivot)))
                #list_ax[i].set_xticklabels(df_pivot['Hari'])
                plt.title(columns_df2[x])
                #plt.xticks(rotation=90)
            plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
            figure.suptitle(year[i], fontsize=16, y=1.03)
            plt.tight_layout()
            plt.show()
        if save == True:
            plt.savefig('tes2.png')
            
        return
histogram_all(DA_I.all_df_pivot, 'all', save=False)        
```

```python
#Heatmap
def heatmap_all(data_frame, date, save=False):
    if date == '2018':
        figure_o = plt.figure(figsize=(20, 10))
        c= data_frame[date][['Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah', 'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir','Minyak Goreng', 'Telur Ayam']].corr()
        ax = sns.heatmap(c,cmap="BrBG",annot=True)
        ax.set_title('Heatmap 2018')
        #figure_o.suptitle(list_daerah2[i], fontsize=30)
        #figure_o.subplots_adjust(top=0.91)
        if save == True:
            plt.savefig('E:\\for github\harga pangan indo 2019\savefig\\'+'jambi_heatmap_2018.png')
        plt.show()
        return
    if date == '2019':
        figure_o = plt.figure(figsize=(20, 10))
        c= data_frame[date][['Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah', 'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir','Minyak Goreng', 'Telur Ayam']].corr()
        ax = sns.heatmap(c,cmap="BrBG",annot=True)
        ax.set_title('Heatmap 2019')
        #figure_o.subplots_adjust(top=0.91)
        if save == True:
            plt.savefig('E:\\for github\harga pangan indo 2019\savefig\\'+'jambi_heatmap_2019.png')
        plt.show()
        return
    year = ['2018', '2019']
    if date == 'all':
        for i in range(len(year)):
            figure_o = plt.figure(figsize=(20, 10))
            c= data_frame[year[i]][['Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah', 'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir','Minyak Goreng', 'Telur Ayam']].corr()
            ax = sns.heatmap(c,cmap="BrBG",annot=True)
            ax.set_title('Heatmap ' + year[i])
            #figure_o.subplots_adjust(top=0.91)
            if save == True:
                plt.savefig('E:\\for github\harga pangan indo 2019\savefig\\'+'jambi_heatmap_' +year[i]+'.png')
            plt.show()
        return
heatmap_all(DA_I.all_df_pivot, date='all', save=False)        
```

```python
#Line PLot
def line_plot_all(data_frame, date, save=False):
    list_ax = [0, 'ax1', 'ax2', 'ax3', 'ax4', 'ax5', 'ax6', 'ax7', 'ax8', 'ax9', 'ax10']
    columns_df = [0, 'Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah',
       'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir',
       'Minyak Goreng', 'Telur Ayam']
    if date == '2018':
        fig = plt.figure(figsize=(20, 10))
        for i in range(1, 11):
            list_ax[i] = plt.subplot(2, 5, i)
            plt.plot(range(len(data_frame[date])), data_frame[date][columns_df[i]])
            #list_ax[i].set_xticks(range(52))
            #list_ax[i].set_xticklabels(range(1, 13))
            plt.ylabel('Harga (Rp)')
            plt.title(columns_df[i])
            plt.xticks(rotation=90)
            plt.xlabel('Jan - Des '+date)
            plt.xticks([])
        plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
        plt.tight_layout()
        fig.suptitle(date, fontsize=16, y=1.03)
        if save == True:
            plt.savefig('tes33.png')
        plt.show()
        return
    if date == '2019':
        fig = plt.figure(figsize=(20, 10))
        for i in range(1, 11):
            list_ax[i] = plt.subplot(2, 5, i)
            plt.plot(range(len(data_frame[date])), data_frame[date][columns_df[i]])
            #list_ax[i].set_xticks(range(52))
            #list_ax[i].set_xticklabels(range(1, 13))
            plt.ylabel('Harga (Rp)')
            plt.title(columns_df[i])
            plt.xticks(rotation=90)
            plt.xlabel('Jan - Des '+date)
            plt.xticks([])
        plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
        plt.tight_layout()
        fig.suptitle(date, fontsize=16, y=1.03)
        if save == True:
            plt.savefig('tes33.png')
        plt.show()
        return
    year = ['2018', '2019']
    if date == 'all':
        for x in range(len(year)):
            fig = plt.figure(figsize=(20, 10))
            for i in range(1, 11):
                list_ax[i] = plt.subplot(2, 5, i)
                plt.plot(range(len(data_frame[year[x]])), data_frame[year[x]][columns_df[i]])
                #list_ax[i].set_xticks(range(52))
                #list_ax[i].set_xticklabels(range(1, 13))
                plt.ylabel('Harga (Rp)')
                plt.title(columns_df[i])
                plt.xticks(rotation=90)
                plt.xlabel('Jan - Des '+year[x])
                plt.xticks([])
            plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
            plt.tight_layout()
            fig.suptitle(year[x], fontsize=16, y=1.03)
            plt.show()
        if save == True:
            plt.savefig('tes33.png')
        return
line_plot_all(DA_I.all_df_pivot, 'all', save=False)
```

```python
#bar plot
def bar_plot(data_frame, date, save=False):
    if date == '2018':
        f, ax = plt.subplots(figsize=(20, 10)) 
        ax = sns.barplot(data=data_frame[date], x='Komoditas', y='Harga', hue='Tanggal')
        plt.ylabel('Harga (Rp)')
        plt.xlabel('Komoditas (365 hari)')
        ax.get_legend().remove()
        #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        #plt.savefig('Bar.png')
        f.suptitle(date, fontsize=16, y= 0.93)
        plt.show()
        return
    if date == '2019':
        f, ax = plt.subplots(figsize=(20, 10)) 
        ax = sns.barplot(data=data_frame[date], x='Komoditas', y='Harga', hue='Tanggal')
        plt.ylabel('Harga (Rp)')
        plt.xlabel('Komoditas (365 hari)')
        ax.get_legend().remove()
        #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        #plt.savefig('Bar.png')
        f.suptitle(date, fontsize=16, y= 0.93)
        plt.show()
        return
    year = ['2018', '2019']
    if date == 'all':
        for i in range(len(year)):
            f, ax = plt.subplots(figsize=(20, 10)) 
            ax = sns.barplot(data=data_frame[year[i]], x='Komoditas', y='Harga', hue='Tanggal')
            plt.ylabel('Harga (Rp)')
            plt.xlabel('Komoditas (365 hari)')
            ax.get_legend().remove()
            f.suptitle(year[i], fontsize=16, y= 0.93)
            #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
            #plt.savefig('Bar.png')
            plt.show()
        return
bar_plot(DA_I.all_df_melt, 'all', save=False)
```

```python
#boxplot
def box_plot_all(data_frame, date, save=False):
    if date == '2018':
        fig = plt.figure(figsize=(20, 10))
        ax = plt.subplot()
        sns.boxplot(data=data_frame[date], x='Komoditas', y='Harga')
        fig.suptitle(date, fontsize=16, y= 0.93)
        plt.show()
        return
    if date == '2019':
        fig = plt.figure(figsize=(20, 10))
        ax = plt.subplot()
        sns.boxplot(data=data_frame[date], x='Komoditas', y='Harga')
        fig.suptitle(date, fontsize=16, y= 0.93)
        plt.show()
        return
    year = ['2018', '2019']
    if date == 'all':
        for i in range(len(year)):
            fig = plt.figure(figsize=(20, 10))
            ax = plt.subplot()
            sns.boxplot(data=data_frame[year[i]], x='Komoditas', y='Harga')
            fig.suptitle(year[i], fontsize=16, y= 0.93)
            plt.show()
        return
box_plot_all(DA_I.all_df_melt, 'all', save=False)
```
