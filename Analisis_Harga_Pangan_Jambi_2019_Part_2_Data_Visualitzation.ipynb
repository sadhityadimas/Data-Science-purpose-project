```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import Analisa_Harga_Pangan_Jambi_2019_Part_1_Data_Analysis as DA_I
```

Analisa Harga Pangan Tahun 2019 di Jambi (Data Harian Sepanjang Tahun)

```python
#histogram

list_ax2 = [0, 'ax1', 'ax2', 'ax3', 'ax4', 'ax5', 'ax6', 'ax7', 'ax8', 'ax9', 'ax10']
columns_df2 = [0, 'Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah',
       'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir',
       'Minyak Goreng', 'Telur Ayam']
plt.figure(figsize=(20, 10))
for i in range(1, 11):
    list_ax2[i] = plt.subplot(2, 5, i)
    plt.hist(DA_I.df_pivot[columns_df2[i]])#, bins=range(len(df_pivot['Bawang Merah'])))
    #list_ax[i].set_xticks(range(len(df_pivot)))
    #list_ax[i].set_xticklabels(df_pivot['Hari'])
    plt.title(columns_df2[i])
    #plt.xticks(rotation=90)
    
plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
plt.tight_layout()
plt.savefig('tes2.png')
plt.show()
```

```python
#heatmap
def heatmap_all(save=False):
        figure_o = plt.figure(figsize=(20, 10))
        c= DA_I.df_pivot[['Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah', 'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir','Minyak Goreng', 'Telur Ayam']].corr()
        sns.heatmap(c,cmap="BrBG",annot=True)
        #figure_o.suptitle(list_daerah2[i], fontsize=30)
        #figure_o.subplots_adjust(top=0.91)
        if save == True:
            plt.savefig('E:\\for github\harga pangan indo 2019\savefig\\'+'jambi_heatmap.png')
        plt.show()
heatmap_all()
```

```python
#Line PLot 
list_ax = [0, 'ax1', 'ax2', 'ax3', 'ax4', 'ax5', 'ax6', 'ax7', 'ax8', 'ax9', 'ax10']
columns_df = [0, 'Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah',
       'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir',
       'Minyak Goreng', 'Telur Ayam']
plt.figure(figsize=(20, 10))
for i in range(1, 11):
    list_ax[i] = plt.subplot(2, 5, i)
    plt.plot(range(len(DA_I.df_pivot)), DA_I.df_pivot[columns_df[i]])
    #list_ax[i].set_xticks(range(52))
    #list_ax[i].set_xticklabels(range(1, 13))
    plt.ylabel('Harga (Rp)')
    plt.title(columns_df[i])
    plt.xticks(rotation=90)
    plt.xlabel('Jan - Des 2019')
    plt.xticks([])
plt.subplots_adjust(top=0.946, bottom=0.059, left=0.059, right=0.989, hspace=0.228, wspace=0.556)
plt.tight_layout()
plt.savefig('tes33.png')
plt.show()
```
