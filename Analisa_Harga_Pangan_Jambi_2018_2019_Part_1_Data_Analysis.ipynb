```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from scipy.stats import mode 
pd.set_option('display.max_rows', 300)
```

```python
#i put 2018 and 2019 data on separate cell and dont put them inside a dictionary yet
#it's because the first step is data cleaning and both data could have different needs of cleaning
#then after both data has been cleaned, i put the pivoted and melted data frame inside a dictionary separated by year
```

```python

#load data to data frame
#2018
#df = pd.read_excel('jambi 2018.xls')
#df.to_csv('jambi 2018.csv')
df_2018 = pd.read_csv('Jambi 2018.csv')
df_2018 = df_2018.drop(columns='Unnamed: 0')
df_2018 = df_2018.drop(columns='No.')
df_2018.rename(columns={'Komoditas(Rp)' : 'Komoditas'}, inplace=True)
df_2018
```

```python
#loading data and doing necessary cleaning
#2019
df_2019 = pd.read_csv('Jambi 2019.csv')
df_2019  = df_2019.drop(columns='Unnamed: 0')
df_2019  = df_2019.drop(columns='No.')
df_2019 .rename(columns={'Komoditas(Rp)' : 'Komoditas'}, inplace=True)

#print(df[['Komoditas', 'No.']])
#print(df.dtypes)
df_2019  = df_2019.drop([1, 2, 3, 4, 5, 6, 8, 10, 11, 13, 15, 17, 19, 20, 22, 23, 25, 26, 27, 29, 30]).reset_index(drop=True)

#df.drop
df_2019.head()
```

```python
#melting each data frame into new data frame and put them in a dictioary separated by year
all_df_melt = {}
year = ['2018', '2019']
df_all = [df_2018, df_2019]
for i in range(len(year)):
    all_df_melt[year[i]] = pd.melt(frame=df_all[i], id_vars=['Komoditas'], var_name="Tanggal", value_name='Harga')
    #you have to change the dtype of Tanngal to datetime, and remember python use american date format, we 
    #to use indonesia format so dayfirst = True, and datetime object for python pandas always yyyy-mm-dd, so dont worry
    all_df_melt[year[i]]['Tanggal'] = pd.to_datetime(all_df_melt[year[i]]['Tanggal'], dayfirst=True)
    all_df_melt[year[i]]=all_df_melt[year[i]].sort_values(by=['Komoditas', 'Tanggal']).reset_index(drop=True)
    all_df_melt[year[i]]['Harga'] = pd.to_numeric(all_df_melt[year[i]]['Harga'], downcast='integer')
```

```python
#pivoting data frames and put them inside a dictionary separated by year
all_df_pivot = {}
for i in range(len(year)):
    all_df_pivot[year[i]] = all_df_melt[year[i]].pivot(columns='Komoditas', index='Tanggal', values='Harga').reset_index()
    all_df_pivot[year[i]]['Hari'] = all_df_pivot[year[i]]['Tanggal'].dt.day
    #df_pivot.Tanggal.str.split('/', expand=True)[0]
```

```python
#performing simple Descriptive Statistic
all_df_aggregate = {}
year = ['2018', '2019']
for i in range(len(year)):
    df_mean = all_df_melt[year[i]].groupby('Komoditas').Harga.mean().round(2).reset_index()
    df_mean.rename(columns={'Harga': 'Mean'}, inplace=True)
    df_median = all_df_melt[year[i]].groupby('Komoditas').Harga.median().reset_index()
    df_median.rename(columns={'Harga': 'Median'}, inplace=True)
    #df_mode = all_df_melt[year[i]].groupby('Komoditas').apply(lambda x: mode(x)[0][0]).reset_index()
    #df_mode.rename(columns={'Harga': 'Mode'}, inplace=True)
    df_std = all_df_melt[year[i]].groupby('Komoditas').Harga.std().round(2).reset_index()
    df_std.rename(columns={'Harga': 'STD'}, inplace=True)
    df_max = all_df_melt[year[i]].groupby('Komoditas').Harga.max().reset_index()
    df_max.rename(columns={'Harga': 'Max'}, inplace=True)
    df_min = all_df_melt[year[i]].groupby('Komoditas').Harga.min().reset_index()
    df_min.rename(columns={'Harga': 'Min'}, inplace=True)
    #df_agregat = df_melt.groupby('Komoditas').Harga.mean().reset_index()
    all_df_aggregate[year[i]] = df_mean.merge(df_median).merge(df_std).merge(df_max).merge(df_min)
    all_df_aggregate[year[i]]['No.'] = range(1, len(all_df_aggregate[year[i]])+1)
    all_df_aggregate[year[i]] = all_df_aggregate[year[i]][['No.', 'Komoditas', 'Mean', 'Median', 'Max', 'Min', 'STD']]

```
