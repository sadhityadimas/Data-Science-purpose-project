```python
#the provinces list
list_daerah = ['aceh', 'bali', 'banten', 'bengkulu', 'diy', 'dki_jakarta', 'gorontalo', 'jabar', 'jambi', 'jateng', 'jatim', 
              'kalbar', 'kalsel', 'kalteng', 'kaltim', 'kalut', 'kepri', 'kepulauan_bangka_belitung', 'lampung', 'maluku',
               'malut', 'ntb', 'ntt', 'papua', 'papua_barat', 'riau', 'sulbar', 'sulsel', 'sulteng', 'sultra', 'sulut', 
               'sumbar', 'sumsel', 'sumut']
#year list
year = ['2018', '2019']

#commodity list
commodity_all = ['Bawang Merah', 'Bawang Putih', 'Beras', 'Cabai Merah',
       'Cabai Rawit', 'Daging Ayam', 'Daging Sapi', 'Gula Pasir',
       'Minyak Goreng', 'Telur Ayam']
```

```python
#First part, loading the data into dataframe
# we are going to load 2018 and 2019 data from each province 
#so i'm going to put them inside a dictionary separated by year

#this dictionary contains 2 dictionaries of each year
all_data_frame = {}
for i in year:
    df_temp = {}
    for a in range(len(list_daerah)):
        if i == '2018':
            df_temp[list_daerah[a]] = pd.read_excel('E:\\for github\\harga pangan indo 2019\\2018\\'+list_daerah[a]+'.xls')
            df_temp[list_daerah[a]].rename(columns={'Komoditas(Rp)': 'Komoditas'}, inplace=True)
            df_temp[list_daerah[a]] = df_temp[list_daerah[a]].drop(columns='No.')
        all_data_frame[i] = df_temp
        if i == '2019':
            drop_index = [1, 2, 3, 4, 5, 6, 8, 10, 11, 13, 15, 17, 19, 20, 22, 23, 25, 26, 27, 29, 30]
            df_temp[list_daerah[a]] = pd.read_excel('E:\\for github\\harga pangan indo 2019\\'+list_daerah[a]+'.xls')
            df_temp[list_daerah[a]].rename(columns={'Komoditas(Rp)': 'Komoditas'}, inplace=True)
            df_temp[list_daerah[a]] = df_temp[list_daerah[a]].drop(columns='No.')
            df_temp[list_daerah[a]] = df_temp[list_daerah[a]].loc[[0, 7, 9, 12, 14, 16, 18, 21, 24, 28]].reset_index(drop=True)
        all_data_frame[i] = df_temp

#since every data has the same table structure, we only need to examine one
#this is 2018 data from kepri
all_data_frame['2018']['kepri'].head()

#some date has empty value in the column
```

```python
#this is 2019 data from kepri
all_data_frame['2019']['kepri'].head()
#good we succesfuly load every dataframe and put them inside a dictionary
```

```python
#The Second Part, Performing Data Cleaning

#first is the reshaping data for data visualization purpose, easier to use with seaborn later
all_data_frame_melt = {}
for i in year:
    df_temp = {}
    for a in range(len(list_daerah)):
        df_temp[list_daerah[a]] = pd.melt(frame=all_data_frame[i][list_daerah[a]], id_vars=['Komoditas'], var_name='Tanggal', value_name='Harga')
    all_data_frame_melt[i] = df_temp

#examine them
all_data_frame_melt['2018']['jateng'].head()
```

```python
#Okay our dataframe is a little bit inconvenience
#some columns has no value at all, it's just '-'
#this means at a certain Date, the market where the data was taken did not operate 
# we are going to drop the rows that has '-' as its values
for i in year:
    for a in range(len(list_daerah)):
        all_data_frame_melt[i][list_daerah[a]] = all_data_frame_melt[i][list_daerah[a]].drop(all_data_frame_melt[i][list_daerah[a]][all_data_frame_melt[i][list_daerah[a]].Harga == '-'].index)

#examine again       
all_data_frame_melt['2018']['jateng'].head()
#succeed, now we proceed to the next step
```

```python
all_data_frame_melt['2018']['sulsel'].dtypes
#Tanggal values supposed to be datetime object, instead it's object
#we need to change it into datetime object
#and also, all values inside the commodity columns are objects, but it is currency (Rupiah) so we need to change it into integer
for i in year:
    for a in range(len(list_daerah)):
        #this loop changes the Tanggal dtypes to datetime object 
        all_data_frame_melt[i][list_daerah[a]]['Tanggal'] = pd.to_datetime(all_data_frame_melt[i][list_daerah[a]]['Tanggal'], dayfirst=True, utc=False)
        #this loop changes dtypes in Harga column to integer
        for b in range(len(commodity_all)):
            all_data_frame_melt[i][list_daerah[a]]['Harga'] = pd.to_numeric(all_data_frame_melt[i][list_daerah[a]]['Harga'], downcast='integer')
        #this loop sorts the value based on commodity names and the dates
        all_data_frame_melt[i][list_daerah[a]] = all_data_frame_melt[i][list_daerah[a]].sort_values(by=['Komoditas', 'Tanggal']).reset_index(drop=True)
all_data_frame_melt['2018']['sulsel'].dtypes
#success
#since there are no errors in the code when we change the value into integer, it means there are no NAN values
#the only inconvenience is the '-' which we had dropped
```

```python
#check again if it works
all_data_frame_melt['2018']['sultra'].head()
#it works, now we can proceed
```

```python
#now we are going to reshape the data frame again, this time for statistical purpose

#we need to reshape the data frame so the commodity column's values will become the columns head with it respective values


#preparing the dictionary to contains all the reshaped DF
all_data_frame_pivot = {}

#looping through each data frame in all_df_melt and reshape them, them put them in all_data_frame_pivot
for i in year:
    df_temp = {}
    for a in range(len(list_daerah)):
        df_temp[list_daerah[a]] = all_data_frame_melt[i][list_daerah[a]].pivot(columns = 'Komoditas', index='Tanggal', values = 'Harga').reset_index()
    all_data_frame_pivot[i] = df_temp
    #this loops adds date and month columns separately
    for a in range(len(list_daerah)):
        all_data_frame_pivot[i][list_daerah[a]]['Hari'] = all_data_frame_pivot[i][list_daerah[a]]['Tanggal'].dt.day
        all_data_frame_pivot[i][list_daerah[a]]['Bulan'] = all_data_frame_pivot[i][list_daerah[a]]['Tanggal'].dt.month
        all_data_frame_pivot[i][list_daerah[a]]['Tahun'] = all_data_frame_pivot[i][list_daerah[a]]['Tanggal'].dt.year

all_data_frame_pivot['2018']['jateng'].head()
 
```

```python
#For now, Our data frames Structur inside of our dictionaies (both in melted and pivoted) are like this:
#all df > df 2018 & df 2019 > Provinces df > commodity table

#now, i am going to make another pivoted and melted data frame but instead of separated by province,
#it will be separated by commodities
#all df > df 2018 & df 2019 > commodities df > province table

#this is for another purpose of statistical proccess and visualizing them later
#after we compare commodities in each province, we may want to compare each commodity in every province

#i can't find a way to take the same commodity type  from each province and put them in a dataframe directly
#so i;m going to take every commodities series from each province, change them into numpy series
#then put them inside a dictionary
#then later i will create new dataframe for each commodities



#now we begin by creating dictionary for commodities
all_numpy_series_commodity = {}

#now fill it with loops from pivoted data frame
for i in year:
    temp2 = {}
    for a in range(len(commodity_all)):
        temp_dict = {}
        #looping through commodity list
        for b in all_data_frame_pivot[i]:
            #looping through province list
            temp_dict[b] = all_data_frame_pivot[i][b][commodity_all[a]].to_numpy()
    #print(all_dict_names[i] )   
        temp2[commodity_all[a]] = temp_dict
    all_numpy_series_commodity[i] = temp2
#examine them
#all_numpy_series_commodity['2019']['Gula Pasir']['kalsel']
#succeed
```

```python
#now i am going to create new data frames for each commodity and put them in a dictionary separated by year

#this is the 2nd version of list_daerah
list_daerah2 = ['Aceh', 'Bali', 'Banten', 'Bengkulu', 'Daerah Istimewa Yogyakarta', 'DKI Jakarta', 'Gorontalo', 'Jawa Barat', 
                'Jambi', 'Jawa Tengah', 'Jawa Timur', 'Kalimantan Barat', 'Kalimantan Selatan', 'Kalimantan Tengah', 
                'Kalimantan Timur', 'Kalimantan Utara', 'Kepulauan Riau', 'Kepulauan Bangka Belitung', 'Lampung', 'Maluku',
               'Maluku Utara', 'Nusa Tenggara Barat', 'Nusa Tenggara Timur', 'Papua', 'Papua Barat', 'Riau', 
                'Sulawesi Barat', 'Sulawesi Selatan', 'Sulawesi Tengah', 'Sulawesi Tenggara', 'Sulawesi Utara', 
                'Sumatera Barat', 'Sumatera Selatan', 'Sumatera Utara']

#since this dataframe will have the same structure as all_dataframe_melt, i'm going to call it:
all_data_frame_commodity_melt = {}
for x in year:
    temp2 = {}
    for i in range(len(commodity_all)):
        temp_dict = {}
        label = []
        value = []
        tanggal = []
        for a in range(len(all_numpy_series_commodity[x][commodity_all[i]])):
            label = label + ([list_daerah2[a]]*len(all_numpy_series_commodity[x][commodity_all[i]][list_daerah[a]]))
            value.append(all_numpy_series_commodity[x][commodity_all[i]][list_daerah[a]])
            tanggal = tanggal + all_data_frame_pivot[x][list_daerah[a]]['Tanggal'].tolist()
        #temp_dict[all_data_frame_commodity[i][a]] = 
        temp_dict = pd.DataFrame({
                "Provinsi": label,
                "Harga": np.concatenate(value),
                "Tanggal" : tanggal
                })
        temp2[commodity_all[i]] = temp_dict
    all_data_frame_commodity_melt[x] = temp2
#examine
all_data_frame_commodity_melt['2018']['Beras'].head()
#succeed
```

```python
#now, like all_data_frame, i'm going to create another shape of dataframe through pivot

all_data_frame_pivot_commodity = {}
for x in year:
    temp2 = {}
    for i in commodity_all:
        temp2[i] = all_data_frame_commodity_melt[x][i].pivot(columns = 'Provinsi', index='Tanggal', values = 'Harga').reset_index()
        #all_data_frame_pivot[i]['Hari'] = all_data_frame_pivot[i].Tanggal.str.split('/', expand=True)[0]
        temp2[i]['Hari'] = temp2[i]['Tanggal'].dt.day
        #all_data_frame_pivot[i]['Bulan'] = all_data_frame_pivot[i].Tanggal.str.split('/', expand=True)[1]
        temp2[i]['Bulan'] = temp2[i]['Tanggal'].dt.month
    all_data_frame_pivot_commodity[x] = temp2

#examine
all_data_frame_pivot_commodity['2018']['Gula Pasir'].head()
#succeed
#there are NAN now
#just fill NAN with mean
#and then make melt dataframe again, because we can't change the mean from melt df comodities, they only has 3 columns
```

```python
#okay so, from the examined table above, there are NAN values
#how so? well i take every commodities series from each province and grouped them in separate tables
#some market did not operate while some did on certain date. thats why there are nan values now
#now i'm going to detect how many nan are there in each columns on each commodity table
#on second thought, i'm just going to check in one table

#missing_val_count_by_column = (all_data_frame_pivot_commodity['2018']['Gula Pasir'].isnull().sum())
        
#print(missing_val_count_by_column[missing_val_count_by_column > 0])

#it's spread out in every province, this is just from one table, i assume the other will be spread out as well
```

```python
#now i'm just going to change the nan values in each column with its mean

for i in year:
    for a in range(len(commodity_all)):
        for b in range(len(list_daerah2)):
            all_data_frame_pivot_commodity[i][commodity_all[a]] = all_data_frame_pivot_commodity[i][commodity_all[a]].fillna(value={list_daerah2[b]:all_data_frame_pivot_commodity[i][commodity_all[a]][list_daerah2[b]].mean()})
#examine the table below
```

```python
#now i'm going to melt all the pivoted commodity dataframe again
#because previously we can't fill the nan values with its columns mean with the melted data frame
all_data_frame_commodity_melt_fix = {}
for i in year:
    df_temp = {}
    for a in range(len(commodity_all)):
        df_temp[commodity_all[a]] = pd.melt(frame=all_data_frame_pivot_commodity[i][commodity_all[a]], id_vars=['Tanggal'],  value_vars=list_daerah2, var_name='Provinsi', value_name='Harga')
    all_data_frame_commodity_melt_fix[i] = df_temp

all_data_frame_commodity_melt_fix['2018']['Beras'].head()
#succeed
```
